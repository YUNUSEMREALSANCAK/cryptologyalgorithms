```python
import numpy as np

alphabet = "abcçdefgğhıijklmnoöprsştuüvyz"

def create_square(text):
    text_length = len(text)
    side_length = int(text_length ** 0.5) + 1
    matrix = [['' for _ in range(side_length)] for _ in range(side_length)]
    index = 0
    for i in range(side_length):
        for j in range(side_length):
            if index < text_length:
                matrix[i][j] = text[index]
                index += 1
            else:
                matrix[i][j] = alphabet[index % len(alphabet)]
                index += 1
    return matrix

def encrypt(text):
    square = create_square(text)
    encrypted_text = ''
    side_length = len(square)
    for i in range(side_length):
        for j in range(side_length):
            encrypted_text += square[j][i]
    return encrypted_text

def decrypt(encrypted_text):
    square = create_square(encrypted_text)
    decrypted_text = ''
    side_length = len(square)
    for i in range(side_length):
        for j in range(side_length):
            decrypted_text += square[i][j]
    return decrypted_text

düzmetin = "mer habadünya"
string = düzmetin.replace(" ", "")
encrypted_text = encrypt(string)
print("Şifrelenmiş Metin:")
for row in create_square(string):
    print(row)
print("\nŞifrelenmiş Metin:", encrypted_text)
decrypted_text = decrypt(encrypted_text)
print("\nÇözülmüş Metin:", decrypted_text)
```

```python
def shift_cipher(shift_count, plaintext_array):
    alphabet_letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    encrypted = [''] * len(plaintext_array)
    for i, letter in enumerate(plaintext_array):
        letter_to_index = alphabet_letters.find(letter)
        letter_to_index += shift_count
        encrypted[i] = alphabet_letters[letter_to_index % len(alphabet_letters)]
    return encrypted

def permutation_cipher(permutation_ints, plaintext_array):
    encrypted = [''] * len(plaintext_array)
    counter = -1
    for i in range(len(plaintext_array)):
        if i % len(permutation_ints) == 0:
            counter += 1
        location = (permutation_ints[i % len(permutation_ints)] - 1)
        encrypted[i] = plaintext_array[location + counter * len(permutation_ints)]
    return encrypted

shift_one = 17
shift_two = 8
permutation_one = [5, 1, 3, 2, 4]
permutation_two = [3, 4, 5, 1, 2]
plaintext = "HELLOWOLRD"
plaintext_array = list(plaintext)

cipher = shift_cipher(shift_one, plaintext_array)


print(plaintext + " -ŞİFRELENMEMİŞ HALİ")
print(''.join(cipher))
```

Vigenère Cipher

```python
# Türk alfabesi
alphabet = "ABCÇDEFGĞHIİJKLMNOÖPRSŞTUÜVYZ"

# Anahtar üretme fonksiyonu
def generateKey(string, key):
    key = list(key)
    if len(string) == len(key):
        return key
    else:
        for i in range(len(string) - len(key)):
            key.append(key[i % len(key)])
    return "".join(key)

# Şifreleme fonksiyonu
def cipherText(string, key):
    cipher_text = []
    for i in range(len(string)):
        x = (alphabet.index(string[i]) +
             alphabet.index(key[i])) % len(alphabet)
        cipher_text.append(alphabet[x])
    return "".join(cipher_text)

# Deşifreleme fonksiyonu
def originalText(cipher_text, key):
    orig_text = []
    for i in range(len(cipher_text)):
        x = (alphabet.index(cipher_text[i]) -
             alphabet.index(key[i]) + len(alphabet)) % len(alphabet)
        orig_text.append(alphabet[x])
    return "".join(orig_text)

if __name__ == "__main__":
    string = "MERHABA DUNYA"
    keyword = "DÖRT"
    string = string.replace(" ", "")
    key = generateKey(string, keyword)
    cipher_text = cipherText(string, key)
    print("Ciphertext:", cipher_text)
    print("Original/Decrypted Text:", originalText(cipher_text, key))
```
